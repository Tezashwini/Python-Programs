```python
# Voltage vs Time
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i)for i in t]
plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.show()
plt.savefig('')
```

```python
# plot Income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv('Data Files/income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3,2:]
plt.title('Income of Arkansas')
plt.xlabel('year')
plt.ylabel('income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# income of all states from 2015 to 2013

incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.show()
```
