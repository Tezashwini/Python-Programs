```python
def secondlargest(li):
    unique=[]
    for n in li:
        if n not in unique:
            unique.append(n)
        unique=sorted(unique,reverse=True)
        return unique[1]
        if len(unique)>1:
            return unique[1]
        return -1
secondlargest([1])
```

```python
def fifthleast(li):
    a=[]
    for i in li:
        if i not in a:
            a.append(i)
    a.sort(reverse=True)
    print(a)
    if len(a)>1:
        return a[4]
    else:
        return -1
fifthleast([1,2,3,4,5])
```

```python
#kth smallest
def ksmallest(li,k):
    unique=[]
    for i in li:
        if i not in unique:
            unique.append(i)
        unique=sorted(unique)
    if len(unique)>=k:
        return unique[k-1]
    return -1
ksmallest([2,3,4,56,7],2)
        
```

```python
def highfrequencyelement(l):
    unique={}
    for n in l:
        if n in unique:
            unique[n]+=1
        else:
            unique[n]=1
    print(unique)
#getting all frequencies into a list
    maxfreq=max(unique.values())
    
#Extract all the keys 
    maxfreqkeys=[]
    for item in unique.items():
        if item[1] == maxfreq:
            maxfreqkeys.append(item[0])
    return min(maxfreqkeys)
highfrequencyelement([1,2,3,2,1,4,4,9])
    
```
