```python
# function to add contact:
import re
fipath='Data Files/Contacts Application.txt'
print("1.Add contacts \n2.list of all Contacts \n3.Search contacts \n4.Delete contacts \n5.Modify contacts ")
while 1:
    k=int(input("Enter 1 for continue \nEnter 0 for exit"))
    if k==0:
        break
    elif k==1:
        print("1.Add contacts \n2.list of all Contacts \n3.Search contacts \n4.Delete contacts \n5.Modify contacts ")
        n=int(input("Enter your choice"))
        if n==1:
            def phonevalidation(no):
                pattern='[9876][0-9]{9}$|^[0][6-9][0-9]{9}$|^[+][9][1][6-9][0-9]{9}$'
                if re.match(pattern,str(no)):
                    return True
                      #print("valid number")
                else:
                    return False
                      #print("invalid number")
                return
            def emailval(email):
                pattern='^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][0-9a-z]{3,18}[.][a-z]{2,4}$'
                if re.match(pattern,email):
                  #print("valid")
                    return True
                else:
                  #print("invalid")
                    return False
            def contactsadding(name,phone,email):
                with open(fipath,'a') as f:
                    if phonevalidation(phone):
                        if emailval(email):       #for lines in f.read().split(','): #if name not in line:    
                            f.write(name+','+str(phone)+','+email+'\n')
                            print("Contact saved successfully") #else:  # print("Contact already existed")
            contactsadding(name,phone,email)
        elif n==2:
           # function to list of all contacts
            def listofallcontacts():
                with open(fipath,'r') as f:
                    for line in f:
                           #print(line)
                        for l in line.split(','):
                            print(l)
            listofallcontacts()
        elif n==3:
               # function to searchcontacts
            def csvToList(fname):
                li = []
                with open(fname, 'r') as f:
                    for line in f:
                        li.append(line.split(','))
                    return li
            def listToFile(li):
                s = ''
                for i in li:
                    s += ','.join(i)
                    return s
            def searchcontact(name):
                fh=csvToList(fipath)
                for line in fh:
                    #print(line)
                    if line[0]==name:
                        print("contact exists")
                        return True
                    print("contact does not exists" )
                    return False
            def searchwithreturn(name):
                   #with open(fipath,'r') as f:
                fh=csvToList(fipath)
                for line in range(len(fh)):
                    if fh[line][0]==name:
                        return fh[line]
                    return -1
                name=input("Enter name")
                print(searchwithreturn(name))
        elif n==4:
            def deletecontact(name):
                res=''
                with open(fipath,'r') as f:
                    res=searchcontact(name)
                    if res==1:
                        fh=csvToList(fipath)
                        for line in range(len(fh)):
                            if fh[line][0]==name:
                                fh.pop(line)
                                s=listToFile(fh)
                                with open(fipath,'w') as f:
                                    f.write(s)
            name=input("Enter name")
            deletecontact(name)
        elif n==5:
            def updatecontact(name,phone):
                res=''
                with open(fipath,'r') as f:
                    res=searchcontact(name)
                    if res==1:
                        fh=csvToList(fipath)
                        for line in range(len(fh)):
                            if fh[line][0]==name:
                                fh[line][1]=str(phone)
                                s=listToFile(fh)
                                with open(fipath,'w') as f:
                                    f.write(s)
            updatecontact(name,phone)
name=input("Enter Contact name")
phone=int(input("Enter phone number"))
email=input("Enter email")
```
